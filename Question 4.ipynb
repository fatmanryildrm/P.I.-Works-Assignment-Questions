```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("country_vaccination_stats.csv")
```

```python
data
```

```python
#countrys with the minimum daily vaccination number

data.groupby("country").agg({"daily_vaccinations": ["min"]})
```

```python
# In this code I fill the missing data with per countrys minimum daily vaccination number

data["daily_vaccinations"] = data["daily_vaccinations"].fillna(data.groupby('country')['daily_vaccinations'].transform('min'))
```

```python
data
```

```python
# Another thing we have to do is if a country does not have any valid vaccination number yet, fill it with “0” (zero)

# For example: I choose Kuwait it has NaN value

data[data["country"].isin(["Kuwait"])]
```

```python
# I fill all NaN values with "0"
data["daily_vaccinations"].fillna(0, inplace = True)
```

```python
# After this code instead of Kuwait's NaN value I filled "0"
data[data["country"].isin(["Kuwait"])]
```
